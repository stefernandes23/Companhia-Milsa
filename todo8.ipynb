##### PARA REALIZAR A ANÁLISE SERÃO NECESSÁRIAS AS IMPORTAÇÕES DAS BIBLIOTECAS: 


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##### INSERÇÃO E VERIFICAÇÃO DE DADOS EM DATAFRAME

---



```python
url = "http://www.leg.ufpr.br/~fernandomayer/data/milsa.csv"
dados = pd.read_csv(url)
dados
```

Coluna funcionário como índice:

```python
indice_func = dados.set_index('Funcionario')
indice_func
```

##### Sabendo que a coluna meses não é necessária para a análise e deve ser retirada, **qual a média de idade dos funcionários?**

###### 1. *RETIRADA DA COLUNA 'MESES' -> .drop()*



```python
retirada_meses = dados.drop(columns=['Meses'])
retirada_meses
```

##### *2. MÉDIA DE IDADE DOS FUNCIONÁRIOS -> mean()*



```python
media_idade_func= dados['Anos'].mean()
media_idade_func
```

Dos funcionários que têm filhos, qual o número mais comum? -> mode()

```python
moda_qtd_filhos = dados['Filhos'].mode()
print('Número mais comum de filhos dos funcionários: ')
moda_qtd_filhos
```

Qual a média e a mediana dos salários dos funcionários? O que esses valores podem indicar?

##### *Mediana*

```python
mediana = dados['Salario'].median()
mediana
```

##### *Média*

```python
media = dados['Salario'].mean()
media
```

Quais são os tipos de instrução existentes? -> .unique()


```python
instrucoes = dados['Inst'].unique()
instrucoes
```

```python
grafico_instrucoes = dados.loc(axis = 1)[['Inst']].value_counts().plot(kind="bar")
grafico_instrucoes.set_title('VARIAÇÃO DE INSTRUÇÃO ENTRE OS FUNCIONÁRIOS DA COMPANHIA MILSA')
grafico_instrucoes.set_ylabel('INSTRUÇÃO DE EDUCAÇÃO')
grafico_instrucoes.set_xlabel('FUNCIONÁRIOS')
```

Qual a média de salário e idade das pessoas casadas?

```python
salario_casados = dados.loc[(dados['Est.civil'] == 'casado'), ['Est.civil', 'Salario']]
salario_casados
```

##### Média Salárial dos Funcionários Casados

```python
media_salarial_casados = salario_casados['Salario'].mean()
media_salarial_casados
```

Média de Idade

```python
idade_casados = dados.loc[(dados['Est.civil']== 'casado'), ['Est.civil', 'Anos']]
idade_casados.mean()
```

Qual o funcionário que possui maior salário (informe seus atributos)?

```python
maior_sal = dados['Salario'].max()
maior_sal
```

Qual o funcionário que possui menor idade (informe seus atributos)?

```python
menor_idade = dados['Anos'].min()
menor_idade
```

Quem são os funcionários que possuem no máximo 35 anos, são da capital e tem instrução de 2o grau?

```python
dados_anos_regiao_inst = dados.loc[(dados['Anos'] <= 35) & (dados['Regiao'] == 'capital') & (dados['Inst'] == '2o Grau')]
dados_anos_regiao_inst
```

Faça uma correlação entre todos os atributos. (Dica: mapeie os atributos qualitativos para atributos quantitativos. Ex: solteiro: 1, casado: 2, etc)

```python
dados['Est.civil'] = dados['Est.civil'].replace(['solteiro', 'casado'], [1,2])
dados['Inst'] = dados['Inst'].replace(['1o Grau', '2o Grau', 'Superior'], [1,2,3])
dados['Regiao'] = dados['Regiao'].replace(['interior', 'capital', 'outro'], [1,2,3])
corr= dados.corr()
corr
```

```python
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, cmap="YlGnBu", annot=True)
```
