##### PARA REALIZAR A ANLISE SERO NECESSRIAS AS IMPORTAES DAS BIBLIOTECAS: 


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##### INSERO E VERIFICAO DE DADOS EM DATAFRAME

---



```python
url = "http://www.leg.ufpr.br/~fernandomayer/data/milsa.csv"
dados = pd.read_csv(url)
dados
```

Coluna funcionrio como ndice:

```python
indice_func = dados.set_index('Funcionario')
indice_func
```

##### Sabendo que a coluna meses no  necessria para a anlise e deve ser retirada, **qual a mdia de idade dos funcionrios?**

###### 1. *RETIRADA DA COLUNA 'MESES' -> .drop()*



```python
retirada_meses = dados.drop(columns=['Meses'])
retirada_meses
```

##### *2. MDIA DE IDADE DOS FUNCIONRIOS -> mean()*



```python
media_idade_func= dados['Anos'].mean()
media_idade_func
```

Dos funcionrios que tm filhos, qual o nmero mais comum? -> mode()

```python
moda_qtd_filhos = dados['Filhos'].mode()
print('Nmero mais comum de filhos dos funcionrios: ')
moda_qtd_filhos
```

Qual a mdia e a mediana dos salrios dos funcionrios? O que esses valores podem indicar?

##### *Mediana*

```python
mediana = dados['Salario'].median()
mediana
```

##### *Mdia*

```python
media = dados['Salario'].mean()
media
```

Quais so os tipos de instruo existentes? -> .unique()


```python
instrucoes = dados['Inst'].unique()
instrucoes
```

```python
grafico_instrucoes = dados.loc(axis = 1)[['Inst']].value_counts().plot(kind="bar")
grafico_instrucoes.set_title('VARIAO DE INSTRUO ENTRE OS FUNCIONRIOS DA COMPANHIA MILSA')
grafico_instrucoes.set_ylabel('INSTRUO DE EDUCAO')
grafico_instrucoes.set_xlabel('FUNCIONRIOS')
```

Qual a mdia de salrio e idade das pessoas casadas?

```python
salario_casados = dados.loc[(dados['Est.civil'] == 'casado'), ['Est.civil', 'Salario']]
salario_casados
```

##### Mdia Salrial dos Funcionrios Casados

```python
media_salarial_casados = salario_casados['Salario'].mean()
media_salarial_casados
```

Mdia de Idade

```python
idade_casados = dados.loc[(dados['Est.civil']== 'casado'), ['Est.civil', 'Anos']]
idade_casados.mean()
```

Qual o funcionrio que possui maior salrio (informe seus atributos)?

```python
maior_sal = dados['Salario'].max()
maior_sal
```

Qual o funcionrio que possui menor idade (informe seus atributos)?

```python
menor_idade = dados['Anos'].min()
menor_idade
```

Quem so os funcionrios que possuem no mximo 35 anos, so da capital e tem instruo de 2o grau?

```python
dados_anos_regiao_inst = dados.loc[(dados['Anos'] <= 35) & (dados['Regiao'] == 'capital') & (dados['Inst'] == '2o Grau')]
dados_anos_regiao_inst
```

Faa uma correlao entre todos os atributos. (Dica: mapeie os atributos qualitativos para atributos quantitativos. Ex: solteiro: 1, casado: 2, etc)

```python
dados['Est.civil'] = dados['Est.civil'].replace(['solteiro', 'casado'], [1,2])
dados['Inst'] = dados['Inst'].replace(['1o Grau', '2o Grau', 'Superior'], [1,2,3])
dados['Regiao'] = dados['Regiao'].replace(['interior', 'capital', 'outro'], [1,2,3])
corr= dados.corr()
corr
```

```python
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, cmap="YlGnBu", annot=True)
```
